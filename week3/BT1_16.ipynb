```python
# Lưu Đức Quang 20205117
# Ngày 21 tháng 10 năm 2021
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Câu a: f0 = 1/50, N = 200, Nbits = 64, 128, 256


```python
N = 200
f0 = 1 / 50
n = np.arange(N)
xn = np.sin(2 * math.pi * f0 * n)

def quantizeandplot(typeofq, nofbit):  #typeofq = 0 nghĩa là truncation, typeofq = 1 nghĩa là rounding
    if typeofq == 0:                        
        z = math.log(2,nofbit)/2
        y = np.arange(-z,z,1)
        delta = 2 / nofbit 
        xqn = (np.trunc(xn*nofbit)/nofbit)  
        en = xqn - xn
        
        plt.subplot(3, 1, 1)
        plt.stem(n, xn, "b-")
        plt.grid(axis = 'y')
        plt.title("USING TRUNCATION Quantization level = %d, x(n)" %nofbit)
        
        plt.subplot(3, 1, 2)
        plt.stem(n, xqn, "m")
        plt.plot(n, xn, "b-")               
        plt.grid(axis = 'y')
        plt.title("USING TRUNCATION Quantization level = %d, $x_q$(n)" %nofbit)
        
        plt.subplot(3, 1, 3)
        plt.stem(n, en, "m")
        plt.grid(axis = 'y')
        plt.title("USING TRUNCATION Quantization level = %d, e(n)" %nofbit)
        
        P_Q = 1/(3*math.pow(4,nofbit))
        P_X = 1 / 2
        sqnr = P_X / P_Q
        print("SQNR =",sqnr)
    if typeofq == 1:
        z = math.log(2,nofbit)/2
        y = np.arange(-z,z,1)
        delta = 2 / nofbit 
        xqn = (np.round(xn*nofbit)/nofbit)  
        en = xqn - xn
        
        plt.subplot(3, 1, 1)
        plt.stem(n, xn, "b-")
        plt.grid(axis = 'y')
        plt.title("USING TRUNCATION Quantization level = %d, x(n)" %nofbit)
        
        plt.subplot(3, 1, 2)
        plt.grid(axis = 'y')
        plt.stem(n, xqn, "m")
        plt.plot(n, xn, "b-")               
        plt.title("USING TRUNCATION Quantization level = %d, $x_q$(n)" %nofbit)
        
        plt.subplot(3, 1, 3)
        plt.stem(n, en, "m")
        plt.grid(axis = 'y')
        plt.title("USING TRUNCATION Quantization level = %d, e(n)" %nofbit)
        
        P_Q = 1/(3*math.pow(4,nofbit))
        P_X = 1 / 2
        sqnr = P_X / P_Q
        print("SQNR =",sqnr)  
```

```python
plt.figure(figsize=(16, 12))
quantizeandplot(0, 64)
```

```python
plt.figure(figsize=(16, 12))
quantizeandplot(0, 128)
```

```python
plt.figure(figsize=(16, 12))
quantizeandplot(0, 256)
```
