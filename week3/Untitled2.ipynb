```python
#Đồng Gang Thép
import numpy as np
import math
import matplotlib.pyplot as plt

N = 50
f0 = 1 / 50
n = np.arange(N)
xn = np.sin(2 * math.pi * f0 * n)
P=10000

plt.figure(figsize=(16, 12))
def quantizeandplot(typeofq, nofbit):  
    if typeofq == 0:                       
        k=(nofbit-1)*xn
        k=np.floor(k)
        if k.any()>0:
            k=k-1
        if k.any()<0:
            k=k+1
        k=k/(nofbit-1)
        en = k - xn
        
        plt.subplot(3, 1, 1)
        plt.stem(n, xn, "b-")
        plt.title("USING TRUNCATION Quantization level = %d, x(n)" %nofbit)
        
        plt.subplot(3, 1, 2)
        s = np.linspace(0, 50, num=P)
        s = s.astype(int)
        xk = np.sin(2 * math.pi * f0 * s)
        s = np.linspace(0, 50, num=P)
        plt.plot(n, xn, "go") 
        plt.plot(s, xk)
        plt.title("USING TRUNCATION Quantization level = %d, $x_q$(n)" %nofbit)
        
        plt.subplot(3, 1, 3)
        plt.stem(n, en, "m")
        plt.title("USING TRUNCATION Quantization level = %d, e(n)" %nofbit)
        
        P_Q = np.sum(en*en) / N
        P_X = 1 / 2
        sqnr = P_X / P_Q
        print("SQNR =",sqnr)
        
```

```python
plt.figure(figsize=(16, 12))
quantizeandplot(0, 64)
```
