# a. N = 100, Fs = 5kHz, Fa = 0.5, 2, 3, 4.5 kHz

```python
def aliasing(Fa1, Fa2, Fs, N):    
    T = N / Fs
    t = np.arange(0, T, 1e-5)
    n = np.arange(N + 1)
    xn1 = np.sin(2 * math.pi * Fa1 * n / Fs)
    xt1 = np.sin(2 * math.pi * Fa1 * t)

    xn2 = np.sin(2 * math.pi * Fa2 * n / Fs + math.pi)
    xt2 = np.sin(2 * math.pi * Fa2 * t + math.pi)

    tn = n / Fs
    #plt.figure(figsize=(30, 20), dpi=200)
    plt.figure(figsize=(12, 12))

    plt.subplot(2,1,1)
    plt.stem(tn, xn1, 'm')
    plt.plot(t, xt1, 'b-')
    plt.title("$F_a$ = %d, $F_s$ = %d" %(Fa1, Fs))

    plt.subplot(2,1,2)
    plt.stem(tn, xn2, 'm')
    plt.plot(t, xt2, 'b-')
    plt.title("$F_a$ = %d, $F_s$ = %d" %(Fa2, Fs))

    plt.show()

Fa1 = 500
Fa2 = 4500
Fs = 5000
N = 100
aliasing(Fa1, Fa2, Fs, N)
aliasing(Fa1, Fa2, Fs, 12)
```

```python
Fa1 = 2000
Fa2 = 3000
Fs = 5000
N = 100
aliasing(Fa1, Fa2, Fs, N)
aliasing(Fa1, Fa2, Fs, 12)
```

# Câu b. Fa = 2 kHz, Fs = 50 kHz
1. Vẽ x(n), xác định fd
2. y(n): even-numbered samples of x(n). Xác định tần số của y(n) và y(t)

```python
Fa = 2000
Fs = 50000
N = 26

T = N / Fs
t = np.arange(0, T, 1e-5)
n = np.arange(N + 1)
xn = np.sin(2 * math.pi * Fa * n / Fs)
xt = np.sin(2 * math.pi * Fa * t)

tn = n / Fs
#plt.figure(figsize=(30, 20), dpi=200)
plt.figure(figsize=(12, 6))

plt.stem(tn, xn, 'm')
plt.plot(t, xt, 'b-')
plt.title("$F_a$ = %d, $F_s$ = %d" %(Fa, Fs))

plt.show()
```

```python
yn = xn[::2]
print(xn.shape)
print(yn.shape)
n2 = np.arange(len(yn))
Ts2 = 2 / Fs
T2 = len(yn) * Ts2
print("T2: ", T2)
t2 = np.arange(0, T2, 1e-5)
xt2 = np.sin(2 * math.pi * Fa * t2)

n2 = n2 * Ts2

plt.figure(figsize=(12, 12))
plt.subplot(2,1,1)
plt.stem(n2, yn, 'm')
plt.plot(t2, xt2, 'b-')
plt.title("$F_a$ = %d, $F_s$ = %d" %(Fa, Fs/2))

plt.subplot(2,1,2)
plt.stem(tn, xn, 'm')
plt.plot(t, xt, 'b-')

plt.show()
```
