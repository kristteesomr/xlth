```python
#Lưu Đức Quang 20205117
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style  
```

```python
Fa = 500
Fs = 5000
N = 100
n = np.arange(N)
print(n)
```

```python
print("a, Vẽ đồ thị xn với 0<=n<=99")
print("xn=sin2pi*Fa/Fs*n")
print("Fs=5kHz") 
print("Fa=0,5;2;3;4,5kHz")
```

```python
xn = np.sin(2*math.pi*Fa/Fs*n)
```

```python
Fa = 500
xn = np.sin(2*math.pi*Fa/Fs*n)
plt.plot(n,xn)
plt.style.use('Solarize_Light2')
plt.title("Đồ thị tín hiệu xn với Fa=500hz")
plt.xlabel("Trục thời gian")
plt.show()
```

```python
Fa = 2000
xn = np.sin(2*math.pi*Fa/Fs*n)
plt.plot(n,xn)
plt.style.use('Solarize_Light2')
plt.title("Đồ thị tín hiệu xn với Fa=2000hz")
plt.xlabel("Trục thời gian")
plt.show()
```
